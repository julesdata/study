## Reshaping Data_Arranging
### sort_values
기능: 값으로 정렬  

**parameters**
* by : 축으로 지정할 column|Index
* axis=0 : 값 비교 단위 (행(0)|열(1) 설정
* ascending=True : 정렬방법 - 오름차순 / 내림차순 (False)
* inplace=False : 정렬결과 원본 바로 적용 여부 
* na_position='last': 결측치 순서 설정

````python
df.sort_values?
````

````python
df.sort_values('mpg') #mpg 컬럼 안의 값의 오름차순으로 정렬
````

````python
df.sort_values('mpg', ascending=False) #내림차순 정렬
````

### sort_index

## Reshaping Data_Tidy Data (Layout chaning)

### melt
: Column을 Row로 gather  

```python
pd.melt(
    frame: 'DataFrame',
    id_vars=None,                         # 기준 열
    value_vars=None,                     #melt할 열
    var_name=None,            #value_vars 목록 이름
    value_name='value',  # value_vars별로 나타낼 값
    col_level=None,
    ignore_index: bool = True,
) -> 'DataFrame
```

````python
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})
df
````

````python
pd.melt(df, id_vars=['A'], value_vars=['B'])
````

````python
pd.melt(df, id_vars=['A'], value_vars=['B','C'])
````

````python
pd.melt(df, value_vars=['A','B','C'])
````

````python
pd.melt(df, value_vars=['A','B','C']).rename(columns={
'variabe' : 'var',
'value': 'val'})
````

### pivot
: row값을 column으로 spread
```python
df.pivot(index=None,   #기준 index로 삼을 column
         columns=None, # row값을 columns화 할 column
         values=None   # 나타낼 값
) -> 'DataFrame'
```

````python
df2 = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two','two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df2
````

````python
df2.pivot(index='foo', columns='bar', values='baz')
````

````python
df3=df2.pivot(index='foo', columns='bar', values='baz').reset_index()
df3
````

### 다시 원래데이터로 녹이기

````python
df3.melt(
    id_vars='foo', 
    value_vars=['A','B','C']).sort_values(
    ['foo','bar']).rename(columns = {'value':'baz'})
````
