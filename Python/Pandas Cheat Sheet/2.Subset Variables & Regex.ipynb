### regex 정규표현식으로 검색하여 가져오기 

**정규표현식**
* '\\.' : Matches strings containing a period '.'
* 'Length\$' : Matches strings ending with word 'Length'
* '^Sepal' : Matches strings beginning with the word 'Sepal'
* '^x[1-5]\$' : Matches strings beginning with 'x' and ending with 1,2,3,4,5
* '^(?!Species$).*' : Matches strings except the string 'Species'

```python
df.filter(regex='_')  #'_'를 포함하는 열이름
```

```python
df.filter(regex='length$') #'lenght'로 끝나는 열이름
```

```python
df.filter(regex='^sepal') #sepal로 시작하는 열이름
```

```python
df.filter(regex='^s.*[a-z]$') #s로 시작하고 알파벳 소문자로 끝나는 열 이름
```

```python
df.filter(regex='^(?!species$).*') #sepcies가 포함되지 않은 열 이름
```

### slicing으로 가져오기

```python
df.loc[2:5,'sepal_length':'petal_width']  #이름으로 가져오기. 모두 포함
```

```python
df.iloc[:,1:3] #위치로 가져오기. 뒷 숫자 불포함
```
